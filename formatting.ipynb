Formatting Retweets,Likes,Views,Comments and Date

```python

def convert_numbers(num_str):
    num_str = str(num_str)  
    try:
        # Handle 'K' for thousands
        if 'K' in num_str:
            return int(float(num_str.replace('K', '').strip()) * 1000)
        # Handle 'M' for millions 
        elif 'M' in num_str:
            return int(float(num_str.replace('M', '').strip()) * 1000000)
        # Handle regular float numbers
        else:
            return int(float(num_str.strip()))  
    except ValueError:
        return np.nan  


columns_to_convert = ['Retweets', 'Likes', 'Views', 'Comments']
for column in columns_to_convert:
    df[column] = df[column].apply(convert_numbers)




def normalize_date(date_str):
   
    date_formats = [
       
        "%b-%d-%Y",       
        "%b-%d-%y",       
        "%b-%d",            
        "%d-%m-%Y",
        "%d-%b-%Y",
        "%d-%b-%y"
    ]
    
    for fmt in date_formats:
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            continue
    
    return pd.NaT  


df['Date'] = df['Date'].apply(normalize_date)

df['Date'] = df['Date'].dt.strftime('%d-%m-%Y')

df

```
