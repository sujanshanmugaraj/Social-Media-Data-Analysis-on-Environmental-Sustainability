```python
import pandas as pd
import random

random.seed(42)


df = pd.read_csv("/content/snda_final.csv")

df['followers_count'] = df['Retweets'].apply(lambda x: random.randint(1000, 10000) + x * 10)


def select_retweet_handle(row):
    sentiment = row['sentiment']
    handle = row['Handle']
    followers_count = row['followers_count']


    filtered_handles = df[(df['sentiment'] == sentiment) & (df['followers_count'] < followers_count) & (df['Handle'] != handle)]

    if filtered_handles.empty:
        return 'none'


    try:
        return random.choice(filtered_handles['Handle'].tolist())
    except IndexError:
        return 'none'



df['retweeted_by'] = df.apply(select_retweet_handle, axis=1)


def select_mutual_friend(row):
    name = row['Name']
    sentiment = row['sentiment']


    if random.random() < 0.1:
        return 'none'


    filtered_friends = df[(df['sentiment'] == sentiment) & (df['Name'] != name)]


    if filtered_friends.empty:
        return 'none'

    return random.choice(filtered_friends['Name'].tolist())

df['mutual_friends'] = df.apply(select_mutual_friend, axis=1)

df.drop('Unnamed: 0', axis=1,inplace=True)

df.to_csv('snda_final_simulation.csv', index=False)

print(df.head())
```

```python
import networkx as nx
import matplotlib.pyplot as plt


B = nx.Graph()

sentiments = df['sentiment'].unique()
handles = df['Handle'].unique()

B.add_nodes_from(handles, bipartite=0)
B.add_nodes_from(sentiments, bipartite=1)
for index, row in df.iterrows():
    B.add_edge(row['Handle'], row['sentiment'])
handle_nodes = {n for n, d in B.nodes(data=True) if d["bipartite"] == 0}
sentiment_nodes = set(B) - handle_nodes


pos = nx.bipartite_layout(B, handle_nodes)


plt.figure(figsize=(10, 8))
nx.draw(B, pos, with_labels=True, node_color=["lightblue" if n in handle_nodes else "blue" for n in B.nodes()], node_size=3000, font_size=6,font_weight='bold', edge_color="gray")
plt.title("Bipartite Graph: Handles and Sentiment Labels", size=15)
plt.show()
```

# STRONG TIES AND WEAK TIES

```python
strong_tie_threshold = 2

strong_ties = []
weak_ties = []


for u, v in G.to_undirected().edges:

    common_neighbors = len(list(nx.common_neighbors(G.to_undirected(), u, v)))


    if common_neighbors >= strong_tie_threshold:
        strong_ties.append((u, v, common_neighbors))
    else:
        weak_ties.append((u, v, common_neighbors))

print("Strong Ties (>= threshold common neighbors):")
for u, v, count in strong_ties:
    print(f"({u}, {v}) - Common Neighbors: {count}")

print("\nWeak Ties (< threshold common neighbors):")
for u, v, count in weak_ties:
    print(f"({u}, {v}) - Common Neighbors: {count}")
```

# TRIADIC CLOSURE

```python
triangle_dict = nx.triangles(G_undirected)
closed_triads = sum(triangle_dict.values()) / 3

open_triads = 0
for node in G_undirected.nodes():
    neighbors = set(G_undirected.neighbors(node))
    for neighbor1 in neighbors:
        for neighbor2 in neighbors:
            if neighbor1 < neighbor2 and not G_undirected.has_edge(neighbor1, neighbor2):
                open_triads += 1

total_triads = open_triads + closed_triads
triadic_closure_ratio = closed_triads / total_triads if total_triads > 0 else 0
print(f"Triadic Closure Ratio: {triadic_closure_ratio:.4f}")
```

*If two users have a common friend then how likely it is for the users to be friends. Here the triadic closure ratio is low indicating less possibility.*

# COMMUNITY DETECTION USING K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

adjacency_matrix = nx.to_numpy_array(G_undirected)

k = 3

kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(adjacency_matrix)
labels = kmeans.labels_

communities = {}
for user_id, community in enumerate(labels):
    user_name = df.loc[user_id, 'Name']
    if community not in communities:
        communities[community] = []
    communities[community].append(user_name)

for community, users in communities.items():
    print(f"Community {community}: {users}")
```

```python
community_colors = [labels[node_id] for node_id in range(len(labels))]

plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G_undirected)

nx.draw(G_undirected, pos, with_labels=True, node_color=community_colors, node_size=300, cmap=plt.cm.Set3, font_size=8, font_weight='bold', edge_color="gray")

plt.title("User Friendship Graph Differentiated by Communities", size=15)
plt.show()
```
